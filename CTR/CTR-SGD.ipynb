```python
def F_selectcategories(df_test, df_train, Categorical_features, frac_min = 0.01):
    selected_categ = {}
    num_selcats = {}
    ntot = df_test.shape[0]
    
    # only the id and ips 0 are common, so I do this categories manually since they take 100*time the others
    #specialfeats = ['device_id', 'device_ip']
    #for feat in specialfeats:
     #   Categorical_features.remove(feat)
      #  selected_categ[feat] = [ 0 ]
       # num_selcats[feat] = 1
        
    for col in Categorical_features:
        # first, the categories that appear in both sets are selected
        # (we want them in test set, to be useful, and in train set, to be trained)
        AllCats_df_test = pd.value_counts(df_test[col])   # category and numcounts are stored (.index and .values)
        AllCats_df_train = df_train[col].unique()  # only need the list of categories
        # ugly array of arrays:
        Cats_testtrain = [[cat,count] for cat,count in zip(AllCats_df_test.index,AllCats_df_test.values)
                               if cat in AllCats_df_train]
        # array of categories, array of counts:
        Cat_arr = [row[0] for row in Cats_testtrain]
        Count_arr = np.array([row[1] for row in Cats_testtrain])

        
        # now we calculate the fraction of the test set that each cat represents, and we
        # select the index of the last category with at least fracmin:
        fracs = Count_arr.astype(np.float)/ ntot

        imax = len(fracs) - 1  # just for the case when the if condition below is never satisfied
        for i in xrange(len(fracs)):
            if fracs[i] < frac_min:
                imax = i-1
                break
        num_selcats[col] = imax+1
                
        # if all categories are represented, remove one of them:
        if num_selcats[col] == len(fracs):
            num_selcats[col] = num_selcats[col]-1
        # selects the categories:
        selected_categ[col] = Cat_arr[0:num_selcats[col]]
    
    # the removed features should be re-included:
    #Categorical_features.extend(specialfeats)
                        
    return selected_categ, num_selcats
```

```python
print 'Selecting categories...:' ; t0 = time()
selected_categ, num_selcats = F_selectcategories(df_test, df_train, Categorical_features, frac_min = frac_min)
print("...done in %0.3fs" % (time() - t0))
```

```python
# New column hourlyclickrate:
print 'New column hourlyclickrate...' ; t0 = time()
hourlyclickrate = (df_train[df_train['click']==1][['click', 'hour']].groupby(['hour']).count() /
                   df_train[['click', 'hour']].groupby(['hour']).count()).values
nameshcrcols = ['hcr']
df_train['hcr'] = df_train['hour'].apply(lambda x: hourlyclickrate[x][0]).astype(np.float16)
df_test['hcr'] = df_test['hour'].apply(lambda x: hourlyclickrate[x][0]).astype(np.float16)
print("...done in %0.3fs" % (time() - t0))
```

```python
def F_new_hcr_col(Secondfeat):
    cols4hcr = ['hour'] #, 'site_category']
    cols4hcr.append(Secondfeat)
    namehcrcol = 'm_' + cols4hcr[0] + '_' + cols4hcr[1]
    print ('    new col %s will be the mean click rate from features: %s' % (namehcrcol, cols4hcr))

    #df_grouped = (df_train[df_train['click']==1].groupby(cols4hcr)['click'].count() /
    #              df_train.groupby(cols4hcr)['click'].count())
    df_grouped = df_train.groupby(cols4hcr)['click'].mean()

    # 1)
    #df_train['hcr'] = df_train[cols4hcr].apply(lambda row: df_grouped.xs(row[0])[row[1]] if row[1] in 
    #                                           selected_categ[cols4hcr[1]] else hourlyclickrate[row[0]][0], axis=1).astype(np.float16)
    #df_test['hcr'] = df_test[cols4hcr].apply(lambda row: df_grouped.xs(row[0])[row[1]] if ((row[1] in 
    #                                         selected_categ[cols4hcr[1]]) and (row[1] in df_grouped.xs(row[0]).index)) 
    #                                         else hourlyclickrate[row[0]][0], axis=1).astype(np.float16)
    # 2)
    # First, a dictionary is created with the mean click rate grouped by hour and second feature:
    hcr2 = {}
    for hour in df_train['hour'].unique():
        hcr2[hour] = {}
        for cat in list(set(list(df_train[cols4hcr[1]].unique()) + list(df_test[cols4hcr[1]].unique()))):
            hcr2[hour][cat] = df_grouped.xs(hour)[cat] if ((cat in selected_categ[cols4hcr[1]]) and 
                                                           (cat in df_grouped.xs(hour).index)) else hourlyclickrate[hour][0]
            #print hour, cat, hcr2[hour][cat]
    # Then, this dict is applied to obtain the new column:
    df_train[namehcrcol] = df_train[cols4hcr].apply(lambda row: hcr2[row[cols4hcr[0]]][row[cols4hcr[1]]], axis=1).astype(np.float16) # raw=True raises error
    df_test[namehcrcol] = df_test[cols4hcr].apply(lambda row: hcr2[row[cols4hcr[0]]][row[cols4hcr[1]]], axis=1).astype(np.float16)

    return namehcrcol
```

```python
print 'Creating new columns hourlyclickrate using hourly mean from extra features...' ; t0 = time()
#for feat in ['C1', 'site_category']:
for feat in Categorical_features[:5]:
    namehcrcol = F_new_hcr_col(feat)
    nameshcrcols.append(namehcrcol)
print("...done in %0.3fs" % (time() - t0))
```

```python
def F_convert_dftypes2int8_except_col(df, nameshcrcols):
    s_temp = df[nameshcrcols]
    for colname in nameshcrcols:
        del df[colname]
    df = df.astype(np.int8, copy=False)
    df[nameshcrcols] = s_temp
    del s_temp
    # the system above is ~50 times faster than the one below.
    #df_test.ix[:,6:] = df_test.ix[:,6:].astype(np.int8, copy=False)
    return df
```

```python
# One-hot encoding et al in chunks:
print 'One-hot process in chunks for train set...'; t0 = time()
num_chunksF = 10
chunkF_size = df_train.shape[0]/num_chunksF
df_train_N = pd.DataFrame()

for i in xrange(num_chunksF):
    r_i = i*chunkF_size
    r_f = (i+1)*chunkF_size
    df_train_Nchunk = pd.get_dummies(df_train.iloc[r_i:r_f,:], columns=Categorical_features, prefix=Categorical_features)
    df_train_Nchunk = F_convert_dftypes2int8_except_col(df_train_Nchunk, nameshcrcols)
    assert set(df_test.columns.tolist()) == set(df_train_Nchunk.columns.tolist())
    df_train_Nchunk = df_train_Nchunk[df_test.columns.tolist()]
    df_train_N = pd.concat([df_train_N, df_train_Nchunk])
df_train = df_train_N
del df_train_N, df_train_Nchunk
print("...done in %0.3fs" % (time() - t0))
```

```python
print 'Preparing M_all and X_sub matrices:' ; t0 = time()
# preparing X, y matrices for CV (for fit we will use chuncks of it)
#X_tot = df_train.values[:,2:] # excluding the feature to be predicted ('click') and 'day'
#y_tot = df_train.values[:,0]
M_all = df_train.values
# Convert test set to array for sklearn predict:
X_sub = df_test.values[:,2:] # .astype(np.int8) excluding the feature to be predicted ('click') and 'day'
print("...done in %0.3fs" % (time() - t0))
del df_train, df_test
```

```python
print 'Shuffling M_all split in train and test sets:' ; t0 = time()
# We split the total matrix M_all into train and test datasets (skl_cv.train_test_split) does not suits my needs because I need the X,y separation afterwards
np.random.shuffle(M_all)
M_all_train = M_all[:0.9*len(M_all),:]
M_all_test = M_all[0.9*len(M_all):,:]
del M_all
print("...done in %0.3fs" % (time() - t0))
```

```python
print 'Reading ID column of test set:' ; t0 = time()
if cluster:
    x=pd.read_csv(originaltestcsvfile, usecols=[0], iterator=True, chunksize=1000000, low_memory=False)
    df_testID=pd.concat(x)
else:
    df_testID=pd.read_csv(originaltestcsvfile, usecols=[0], nrows=100000)
print("...done in %0.3fs" % (time() - t0))
```

### Grid of SGD classifiers:

```python
chunk_size = np.shape(M_all_train)[0]/num_chunks
print 'Starting fitting of SGD in chunks with partial_fit...' ; t0f = time()
print '    (there are ',num_chunks,'chunks with ',chunk_size,'rows)'

print '  A grid of learning rates alpha will be tried:'
print "             alphas = ", ("%f "*len(alphas) % tuple(alphas))
cvscore_a = {}
logloss = {}

for alpha in alphas:
    print '** alpha = ',alpha ; t0al = time()
    sgd = skl_lm.SGDClassifier(loss='log', penalty=penalty, alpha=alpha, n_jobs=4, verbose=0)
    # when doing .partial_fit, it only does 1 epoch (no matter which n_iter do you pass, and shuffle is useless as well

    # manually iterating because partial_fit does only one epoch:
    print '  Fitting classifier...'
    for e in xrange(num_epochs):
        print '    Epoch: ',(e+1),'/',num_epochs ; t0e = time()
        #df_train_shuf = df_train.ix[np.random.choice(df_train.index, len(df_train))]
        np.random.shuffle(M_all_train)
        
        for i in xrange(num_chunks):
            #t0 = time()
            r_i = i*chunk_size
            r_f = (i+1)*chunk_size 
            #X = df_train_shuf.values[r_i:r_f,2:] # excluding the feature to be predicted ('click') and 'day'
            #y = df_train_shuf.values[r_i:r_f,0]
            X = M_all_train[r_i:r_f,2:]
            y = M_all_train[r_i:r_f,0]

            sgd.partial_fit(X, y, classes=np.unique(y))
            #print("    ... Chunk %i / %i done in %0.3fs" % ((i+1), num_chunks, time() - t0))
        print("    ... all chunks of epoch %i done in %0.3fs" % ((e+1), time() - t0e))
        print("    ... temp acc: %0.4f  ;  logloss: %0.4f" % (sgd.score(M_all_test[:,2:], M_all_test[:,0]),
                                                              skl_met.log_loss(M_all_test[:,0], sgd.predict_proba(M_all_test[:,2:]))))
    del X, y #df_train_shuf
    print("  ... all epochs for partial fit done in %0.3fs" % (time() - t0al))
    
    # cross_val_score computes CV scores from 3-fold stratified cross-validation procedure:
    #print '  Calculating a score through CV...' ; t0cv = time()
    #np.random.shuffle(M_all)
    #cvscores = skl_cv.cross_val_score(sgd, M_all[:,2:], M_all[:,0], cv=2)
    #cvscore_a[alpha] = cvscores.mean()
    #print("  ... Accuracy: %0.2f (+/- %0.2f), calculated in %f s" % (cvscores.mean(), cvscores.std() * 2, (time() - t0cv)))
    print '  Calculating accuracy on a separated test set, without CV...' ; t0cv = time()
    cvscore_a[alpha] = sgd.score(M_all_test[:,2:], M_all_test[:,0])
    print("  ...  acc = %f, done in %0.3fs" % (cvscore_a[alpha], (time() - t0cv)))
    
    print '  Calculating logaritmic loss on the same separated test set, without CV...' ; t0ll = time()
    logloss[alpha] = skl_met.log_loss(M_all_test[:,0], sgd.predict_proba(M_all_test[:,2:]))
    print("  ...  logloss = %f, done in %0.3fs" % (logloss[alpha], (time() - t0ll)))
    
    print '  Making preditions...' ; t0pred = time()
    y_prob = sgd.predict_proba(X_sub)  # Probability estimates
    print("  ...done in %0.3fs" % (time() - t0pred))
    
    print '  IDs and probs have the same length?'
    assert len(df_testID.id.values) == len(y_prob)
    
    print '  Writing output file with predictions:' ; t0 = time()
    outcsvfile = outcsvfilePATH + 'SGD-f' + str(frac_min) + '-ch' + str(num_chunks) + '-e' + str(num_epochs) + '-log-a' + str(alpha) + '.sub'
    print '  outputfile full path & name:',outcsvfile
    with open(outcsvfile, 'w') as outfile:
        outfile.write('id,click\n')
        for id, prob in zip(df_testID.id.values[:len(y_prob[:,1])], y_prob[:,1]):
            outfile.write('%s,%s\n' % (id, prob))
    print("  ...done in %0.3fs" % (time() - t0))

print ("The fitting of the grid of SGD was done in %0.3fs\n" % (time() - t0f))
print 'The alpha with max accuracy is:', max(cvscore_a.iteritems(), key=operator.itemgetter(1))[0]
print 'The alpha with min logloss is:', min(logloss.iteritems(), key=operator.itemgetter(1))[0]
print ' alpha  accuracy  logloss'
for alpha in alphas:
    print(' %.0e   %0.4f   %0.4f' % (alpha,cvscore_a[alpha],logloss[alpha]))
```
